```r
diamonds1 <- copy(diamonds)
```

```r
setDT(diamonds1)
```

```r
diamonds2 <- diamonds1[, .SD[1], by = c("clarity", "carat")]
```

```r
# Make a pivot_wider operation
diamonds_at_wide <- diamonds2 %>% pivot_wider(id_cols = carat, names_from = clarity, values_from = price)
```

```r
diamonds_at_wide
```

```r
# Make a pivot_longer operation
diamonds_at_long <- diamonds_at_wide %>% pivot_longer(cols = "carat", names_to = "cut", values_to = "price", values_drop_na = T)
```
